### Pandas Visualization Practice

```
import numpy as np
import pandas as pd
%matplotlib inline
# import seaborn as sns #just for better styling
import matplotlib.pyplot as plt 
plt.style.use("seaborn") # ggplot
```

```
df1_url= "https://raw.githubusercontent.com/shivampip/python-for-finance/master/04-Visualization-Matplotlib-Pandas/04-02-Pandas%20Visualization/df1"
df2_url= "https://raw.githubusercontent.com/shivampip/python-for-finance/master/04-Visualization-Matplotlib-Pandas/04-02-Pandas%20Visualization/df2"
df3_url= "https://raw.githubusercontent.com/shivampip/python-for-finance/master/04-Visualization-Matplotlib-Pandas/04-02-Pandas%20Visualization/df3"
```

```
df1= pd.read_csv(df1_url, index_col= 0)
df2= pd.read_csv(df2_url)
df3= pd.read_csv(df3_url)
```

```
df1.head()
```

```
# df1['A'].hist()
df1['A'].hist(bins= 30)
```

```
df1['A'].plot(kind='hist', bins= 40)
```

```
df1['A'].plot.hist()
```

```
df2.plot.area()
```

```
df2.plot.bar()
```

```
df2.plot.bar(stacked= True)
```

```
df1['A'].plot.line(x= df1.index, y= 'A', figsize=(10, 2))
```

```
df1.plot.scatter(x= 'A', y= 'B', c= df1['C']) # 3rd axis is color
```

```
df1.plot.scatter(x= 'A', y= 'B', s= df1['C']*100) # 3rd axis is size 
```

```
df2.plot.box()
```
