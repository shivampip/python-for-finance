## Matplotlib Practice

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
x= np.linspace(0, 100, 100)
y= x ** 2
```

### Functional Method

```
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g')
```

### Object Oriented Method

```
fig= plt.figure()

# [left, bottom, width, height] (all in percentage)
axes= fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("This is Title")
```

```
fig= plt.figure()

axes1= fig.add_axes([0, 0, 1, 1])
axes2= fig.add_axes([0, 0.8, 0.3, 0.2])

axes1.plot(x, y)
axes1.set_title("Bigger Plot")

axes2.plot(y, x)
axes2.set_title("Smaller Plot")
```

```
fig, axes= plt.subplots(2,3)
# plt.tight_layout()
```

```
fig, axes= plt.subplots(1, 3)

# for axe in axes:
#     axe.plot(x, y)

axes[0].plot(x, y)
axes[0].set_title("First Plot")
axes[1].plot(x, x)
axes[2].plot(y, x)
axes[2].set_title("Third Plot")
```

#### Figure size and DPI

```
fig, axes= plt.subplots(2, 2,figsize= (7, 2))

axes[0,0].plot(x, y)
axes[1,1].plot(y, x)
plt.tight_layout()
```

```
# LEGEND
fig, axes= plt.subplots(figsize= (4,4))

xx= np.linspace(1, 5, 10)
axes.plot(xx, xx**2, label= "Square")
axes.plot(xx, xx**3, label= "Cube")

axes.set_title("Title")
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")

axes.legend()
axes.legend(loc=0) # 0= choose best
```
